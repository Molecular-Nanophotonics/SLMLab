```python
import numpy as np
pi = np.pi
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
-128-64
```

```python
np.arange(-256, 255+64, 64)
```

```python
slm_size = 512
x, y = np.meshgrid(np.arange(-256, 255+64, 64), np.arange(-256, 255+64, 64))
x = x.flatten()
y = y.flatten()
h_I = np.ones(81)
```

```python
h_dAmplitude = np.empty(81)

for i in range(81):
    sincx = 1.0 if (x[i] == 0) else pi*x[i]/slm_size/np.sin(pi*x[i]/slm_size)
    sincy = 1.0 if (y[i] == 0) else pi*y[i]/slm_size/np.sin(pi*y[i]/slm_size)
    h_dAmplitude[i] = 1.0 if (h_I[i] <= 0.0) else sincx * sincy * np.sqrt(h_I[i]/100) * slm_size * slm_size
```

```python
h_dAmplitude
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x, y, s=50, c=h_dAmplitude, cmap='viridis')
plt.xlim(-300, 300)
plt.ylim(-300, 300)
plt.colorbar()
```

```python
np.max(h_dAmplitude)
```
